```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# URL of the website to scrape
url = 'https://it.wallapop.com/app/search?longitude=12.4942&latitude=41.8905&filters_source=search_box&keywords=macbook%20air%20m2%2016%2F512&min_sale_price=500&max_sale_price=1000'

# Path to the EdgeDriver executable
chrome_driver_path = r'C:\Users\lucam\Downloads\edgedriver_win32\msedgedriver.exe'

# Set up the Selenium WebDriver for Edge
service = Service(chrome_driver_path)
driver = webdriver.Edge(service=service)

# Open the URL
driver.get(url)

# Wait for the "Rifiuta tutto" button and click it
# try:
#     rifiuta_button = WebDriverWait(driver, 10).until(
#         EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Rifiuta tutto')]"))
#     )
#     rifiuta_button.click()
# except Exception as e:
#     print("Rifiuta tutto button not found or not clickable:", e)

# # Wait for the "salta" button and click it, if it exists
# try:
#     salta_button = WebDriverWait(driver, 5).until(
#         EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'salta')]"))
#     )
#     salta_button.click()
# except Exception as e:
#     print("Salta button not found or not clickable, proceeding without clicking it:", e)

# wait 2 seconds for the page to load
time.sleep(4)


# make a list of ItemCardList__item
items = driver.find_elements(By.XPATH, "//a[@class='ItemCardList__item']")

for item in items:
    print(item.get_attribute('href'))
    # Open the item in a new tab
    driver.execute_script("window.open(arguments[0].href, '_blank');", item)
    driver.switch_to.window(driver.window_handles[-1])
    time.sleep(4)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    # Extract items with 
    # - class="item-detail_ItemDetail__description__7rXXT"
    # - class="item-detail-price_ItemDetailPrice--standard__TxPXr"
    print(soup.find('section', class_='item-detail_ItemDetail__description__7rXXT').text)
    print(soup.find('span', class_='item-detail-price_ItemDetailPrice--standard__TxPXr').text)

    # Close the tab and switch back to the main window
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(2)

# Wait for the page to load
time.sleep(5)  # Adjust the sleep time if necessary

# Get the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')


# Close the WebDriver
driver.quit()
```
